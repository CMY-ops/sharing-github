```python
# 2022 Weekly Net Income Calculator for Hourly Employees working in New York City, claiming 0 dependents. 

# This Code is a work in progres:
# Errors & Exception Handling for user inputs will be perfected soon with the use of loops.
# 2023 tax numbers will soon be switched out for these 2022 numbers.

# Goal: Input Weekly hours worked and Gratuity per hour, then Return Gross Income & other financial percentages.

print("")
print("WELCOME TO CAM'S WEEKLY INCOME TAX CALCULATOR!")
print("")

Total_Hours = float(input("Enter Total Hours Worked, in the following format 00.00:"))

Hourly_Rate = float(input("Enter your hourly pay rate in the following format 00.00:"))

Total_Gratuity = float(input("Enter the gratuity you earned per hour in the following format 00.00:"))



if Total_Hours <= 40.00:
    Hourly_Pay = Total_Hours * Hourly_Rate
elif Total_Hours > 40.00:
    Hourly_Pay = (40*Hourly_Rate) + ((Total_Hours - 40.00) * (Hourly_Rate*1.5))
    

###    
    
# Adding in retirement account withholding percentage option as an input from user,
# Roth contributions are calculated based on gross pay but deducted from net pay.  .... 01/20/2023

retirement_amount = int(input("Enter percentage you'd like withheld for Roth 401K (do not include % sign):"))
retirement_percentage = float(retirement_amount/100)




weekly_retirement_contribution = round((((Total_Hours * Total_Gratuity) + Hourly_Pay)*retirement_percentage),2)

### Added option to define employer contribution percentage, if there is one 02/03/2023

retirement_match = int(input("Please enter your employer's matching contribution percentage (do not include % sign). If there is no match, just put 0:"))

###

weekly_retirement_match = round((((Total_Hours * Total_Gratuity) + Hourly_Pay)*(retirement_match/100)),2)

total_weekly_retirement_savings = weekly_retirement_contribution + weekly_retirement_match


Total_Gross = round(((Total_Hours * Total_Gratuity) + Hourly_Pay),2)

   
    
###



# Goal: Take newfound Gross Income and determine Net Income through various tax brackets

# FEDERAL TAX BRACKET CALCULATIONS:

if Total_Gross < 197.60:
    Fed_Tax = round((Total_Gross * .10),2)
    

elif Total_Gross >= 197.60 and Total_Gross <= 803.36:
    Fed_Tax = round((19.76 + ((Total_Gross -197)* .12)),2)
    
    
elif Total_Gross >= 803.37 and Total_Gross <= 1712.98:
    Fed_Tax = round((92.45 + ((Total_Gross -803.37)* .22)),2)
    
    
elif Total_Gross >= 1712.99 and Total_Gross <= 3270.19:
    Fed_Tax = round((292.57 + ((Total_Gross -1713)* .24)),2)
    
    
elif Total_Gross >= 3270.20 and Total_Gross <= 4152.88:
    Fed_Tax = round((666.30 + ((Total_Gross -3270.21)* .32)),2)
    
    
elif Total_Gross >= 4152.89 and Total_Gross <= 10382.69:
    Fed_Tax = round((948.76 + ((Total_Gross -4152.89)* .35)),2)
    
elif Total_Gross >= 10382.70:
    Fed_Tax = round((3129.19 + ((Total_Gross - 10365.38)* .37)),2)
                    
else:
    print("Unable to Calculate Federal Tax.")
    
Fed_Tax_Percent = round((Fed_Tax/Total_Gross)*100,2)


# NEW YORK STATE TAX BRACKET CALCULATIONS:

if Total_Gross >= 0.00 and Total_Gross <= 163.47:
    State_Tax = round((Total_Gross*.04),2)

elif Total_Gross >= 163.48 and Total_Gross <= 225.00:
    State_Tax = round((6.54 + ((Total_Gross - 163.46)*.045)),2)

elif Total_Gross >= 225.01 and Total_Gross <= 267.31:
    State_Tax = round((9.31 + ((Total_Gross - 225)*.0525)),2)

elif Total_Gross >= 267.32 and Total_Gross <= 411.54:
    State_Tax = round((11.54 + ((Total_Gross - 267.30)*.059)),2)

elif Total_Gross >= 411.55 and Total_Gross <= 1550.96:
    State_Tax = round((20 + ((Total_Gross - 411.55)*.0597)),2)
    
elif Total_Gross >= 1550.97 and Total_Gross <= 4142.30:
    State_Tax = round((88 + ((Total_Gross - 1551)*.0633)),2)

elif Total_Gross >= 4142.31 and Total_Gross <= 20722.11:
    State_Tax = round((252 + ((Total_Gross - 4142.31)*.0685)),2)
    
else:
    print("Unable to Calculate State Tax.")
    
State_Tax_Percent = round((State_Tax/Total_Gross)*100,2)
    

# NEW YORK CITY TAX BRACKETS

if Total_Gross <= 230.76:
    NYC_Tax = round((Total_Gross * .03078),2)

elif Total_Gross >= 230.77 and Total_Gross <= 480.76:
    NYC_Tax = round(7.10 + ((Total_Gross - 230.77)* .03762),2)
    
elif Total_Gross >= 480.77 and Total_Gross <= 961.53:
    NYC_Tax = round(16.50 + ((Total_Gross -480.76)* .03819),2)
    
elif Total_Gross > 961.53:
    NYC_Tax = round(34.87 + ((Total_Gross - 961.53)* .03867),2)
    
else:
    print("Unable to Calculate NYC Tax.")
    
NYC_Tax_Percent = round((NYC_Tax/Total_Gross)*100,2)
    
# SOCIAL SECURITY FLAT TAX

Social_Security_Tax = round((Total_Gross * .062),2)
Social_Security_Tax_Percent = 6.2

# MEDICARE FLAT TAX

Medicare_Tax = round((Total_Gross * .0145),2)
Medicare_Tax_Percent = 1.45

# NY PAID FAMILY LEAVE TAX???

NYPFL_Tax = round((Total_Gross * .00511),2)
NYPFL_Tax_Percent = 0.511


# Did you Earn Spread of Hours? Did you work more than 10 hours in any given shift?
    
try:
    Spread_of_Hours = input("Spread of Hours: Did you work more than 10 hours in any shift this week?: Y or N")
    
    if Spread_of_Hours.upper() == 'Y':
        Spread_Value = input("How many shifts did you work more than 10 hours? Please answer in numerical format.")
        
        Total_Spread_of_Hours_Income = (float(Spread_Value) * 15)
        Total_Gross = Total_Gross + Total_Spread_of_Hours_Income
        
    else:
        Total_Spread_of_Hours_Income = 0
        pass

except:
    print("Spread of Hours Function failed.")
        


# Choose whether or not to add in deductions:

try:
    ### Adding Personalized Deductions option as a number OR as a percent.

    Add_Deductions = input("Would you like to add any other deductions to this calculation? Y or N:")

    if Add_Deductions.upper() == 'Y':
        
        Number_Or_Percentage = input("Please enter 'A' or 'P' to select a Post-Tax Amount or a Pre-Tax Percentage you'd like to be withheld: A or P")
        
        if Number_Or_Percentage.upper() == 'A':
            Deduction_Amount = float(input("Please enter the amount you'd like to be deducted, post-tax, in the following format 00.00:"))

        
        elif Number_Or_Percentage.upper() == 'P':
            Deduction_Percent = (float(input("Please enter the percentage you'd like to be deducted, pre-tax (do not include % sign):"))/100)
            Deduction_Amount = (Deduction_Percent * Total_Gross)
            
            
    elif Add_Deductions.upper() == 'N':
        print("No Deductions were added.")
        pass
    else:
        print("Unable to process entry. Deductions not added to tax witholding.")
        pass
except:
    print("Deductions Process failed.")
    

Deduction_Percentage = round(((Deduction_Amount/Total_Gross)*100),2)         


# ADD UP TOTAL TAXES

# Total_Tax = round((Fed_Tax + State_Tax + Social_Security_Tax + Medicare_Tax +NYPFL_Tax),2)

Total_Tax = round((Fed_Tax + State_Tax + Social_Security_Tax + Medicare_Tax +NYPFL_Tax),2)


# RETURN NET INCOME AFTER TAXES:

Net_Income = round(((Total_Gross - Total_Tax)-weekly_retirement_contribution-Deduction_Amount),2)

Percentage_Taken_In_Taxes = round(((Total_Tax / Total_Gross)*100),2)

Percentage_Income_Kept = round(((Net_Income / Total_Gross)*100),2)


# Print Statements and Display Output:

print(" ")
print("***************************************************************************************************************")
print(" ")

print(" ")
print("Total Gross Income (Hourly Pay + Gratuity):")
print(Total_Gross)

print(" ")
print("Spread of Hours ($15/hr):")
print(Total_Spread_of_Hours_Income)

print(" ")
print("***************************************************************************************************************")
print(" ")

print("Tax Breakdown & Percentages of Total Gross Income:")

print(" ")
print("Federal Tax:")
print("{x} ({y} %)".format(x=Fed_Tax,y=Fed_Tax_Percent))

print("")
print("State Tax:")
print("{x} ({y} %)".format(x=State_Tax,y=State_Tax_Percent))

print(" ")
print("NYC Tax:")
print("{x} ({y} %)".format(x=NYC_Tax,y=NYC_Tax_Percent))

print(" ")
print("Social Security Tax:")
print("{x} ({y} %)".format(x=Social_Security_Tax,y=Social_Security_Tax_Percent))

print(" ")
print("Medicare Tax:")
print("{x} ({y} %)".format(x=Medicare_Tax,y=Medicare_Tax_Percent))

print(" ")
print("New York Paid Family Leave Tax:")
print("{x} ({y} %)".format(x=NYPFL_Tax,y=NYPFL_Tax_Percent))

print(" ")
print("***************************************************************************************************************")
print(" ")

print(" ")
print("Total Taxes Paid:")
print("{x} ({y} %)".format(x=Total_Tax,y=Percentage_Taken_In_Taxes))

print(" ")
print("Total Retirement Contribution:")
print("{x} ({y} %)".format(x=weekly_retirement_contribution,y=retirement_amount))

print(" ")
print("Total Deductions:")
print("{x} ({y} %)".format(x=Deduction_Amount,y=Deduction_Percentage))

print(" ")
print("***************************************************************************************************************")
print(" ")

print(" ")
print("Net Income:")
print("{x} ({y} %)".format(x=Net_Income,y=Percentage_Income_Kept))

print(" ")
print("Weekly Retirement Savings (Your {x}% contribution with {y}% Employer Match):".format(x=retirement_amount,y=retirement_match))
print("{x}".format(x=total_weekly_retirement_savings))
```
